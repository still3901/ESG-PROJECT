```python
import pandas as pd
import numpy as np
import os
```

```python
# df1 = pd.read_csv('emp2022.csv')
df1 = pd.read_excel('/content/drive/MyDrive/KDT/파이널프로젝트/2019_직원 데이터.xlsx')
```

```python
df1 = df1.loc[:, ['corp_code','corp_name','sexdstn','rgllbr_co','sm','avrg_cnwk_sdytrn','jan_salary_am']]
# df2 = df2.loc[:, ['corp_code','corp_name','sexdstn','rgllbr_co','sm','avrg_cnwk_sdytrn','jan_salary_am']]
```

```python
df1
```

```python
# 이상값 대체
df1.replace('-', '', regex =True, inplace=True)
df1.replace('--', '', regex =True, inplace=True)
df1.replace(',', '', regex =True, inplace=True)

# object타입을 float타입으로 변환
df1['rgllbr_co'] = pd.to_numeric(df1['rgllbr_co'])
df1['sm'] = pd.to_numeric(df1['sm'])
df1['jan_salary_am'] = pd.to_numeric(df1['jan_salary_am'])

# 결측치 모두 0값으로 대체
df1.fillna(0, inplace = True)

# 결과 확인
df1.info()

df1.isnull().sum()
```

```python
# avrg_cnwk_sdytrn 제거
df1.drop(columns=['avrg_cnwk_sdytrn'], inplace=True)

# 결과 확인
print(df1)
```

```python
# df1 저장
df1.to_csv('df1.csv', index=False)
```

```python
# 성별(sexdstn), 회사 코드(corp_code), 회사 이름(corp_name)별로 합계 계산
result = df1.groupby(['corp_code', 'corp_name', 'sexdstn']).agg({
    'rgllbr_co': 'sum',
    'sm': 'sum',
    'jan_salary_am': 'sum'
}).reset_index()

print(result)
```

```python
df = result.iloc[:,1:].copy()
df
```

```python
# pivot_table을 사용하여 회사별로 성별을 열로 변환
df_pivot = df.pivot_table(index='corp_name', columns='sexdstn', values=['rgllbr_co', 'sm', 'jan_salary_am'], aggfunc='sum')

# 열 이름 정리
df_pivot.columns = [f'{col}_{sexdstn.lower()}' for col, sexdstn in df_pivot.columns]

df_pivot['total_emp'] = df_pivot['sm_남'] + df_pivot['sm_여']
df_pivot.fillna(0, inplace= True)
df_pivot.isnull().sum()
df_pivot
```

```python
# women_ratio 추가
df_pivot['women_ratio'] = df_pivot['sm_여'] / (df_pivot['total_emp']) * 100

# man_ratio 추가
df_pivot['man_ratio'] = df_pivot['sm_남'] / (df_pivot['total_emp']) * 100

# Full-time employee ratio 추가 (남)
df_pivot['Full-time employee ratio_남'] = df_pivot['rgllbr_co_남'] / df_pivot['sm_남']

# Full-time employee ratio 추가 (여)
df_pivot['Full-time employee ratio_여'] = df_pivot['rgllbr_co_여'] / df_pivot['sm_여']
```

```python
result = df_pivot.copy()
result
```

```python
result.to_csv('emp2019.csv')
```
