```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('/content/drive/MyDrive/KDT/최종프로젝트-연습용/merged_df_06_25_지원.xlsx')
```

```python
# 결측치가 있는 컬럼 이름 조회
columns_with_nan = df.columns[df.isnull().any()].tolist()

print("결측치가 있는 컬럼 이름:", columns_with_nan)
```

```python
# 특정 컬럼의 결측치만 0으로 채우기
df['용수 사용량(Ton)'] = df['용수 사용량(Ton)'].fillna(0)
df['에너지 총량(TJ)'] = df['에너지 총량(TJ)'].fillna(0)
df['온실가스 배출총량(TonCo2eq)'] = df['온실가스 배출총량(TonCo2eq)'].fillna(0)

print("\n특정 컬럼의 결측치만 0으로 채우기:")
print(df)
```

```python
# 용수 사용량 조건 정의
conditions = [
    df['용수 사용량(Ton)'] == 0,
    (df['용수 사용량(Ton)'] > 0) & (df['용수 사용량(Ton)'] < 8916320.316036444),
    (df['용수 사용량(Ton)'] >= 8916320.316036444) & (df['용수 사용량(Ton)'] < 40900794.02807971),
    (df['용수 사용량(Ton)'] >= 40900794.02807971) & (df['용수 사용량(Ton)'] < 108022783.89916667),
    df['용수 사용량(Ton)'] >= 108022783.89916667
]

# 각 조건에 해당하는 값
values = [5, 1, 2, 3, 4]

# 조건에 따라 값을 할당
df['용수 사용량 등급'] = np.select(conditions, values)
```

```python
# 에너지 총량 조건 정의
conditions1 = [
    df['에너지 총량(TJ)'] == 0,
    (df['에너지 총량(TJ)'] > 0) & (df['에너지 총량(TJ)'] < 39577.83371812657),
    (df['에너지 총량(TJ)'] >= 39577.83371812657) & (df['에너지 총량(TJ)'] < 129813.46140846153),
    (df['에너지 총량(TJ)'] >= 129813.46140846153) & (df['에너지 총량(TJ)'] < 301928.6705833333),
    df['에너지 총량(TJ)'] >= 301928.6705833333
]

# 각 조건에 해당하는 값
values1 = [5, 1, 2, 3, 4]

# 조건에 따라 값을 할당
df['에너지 총량 등급'] = np.select(conditions1, values1)
```

```python
# 온실가스 배출총량 조건 정의
conditions2 = [
    df['온실가스 배출총량(TonCo2eq)'] == 0,
    (df['온실가스 배출총량(TonCo2eq)'] > 0) & (df['온실가스 배출총량(TonCo2eq)'] < 39577.83371812657),
    (df['온실가스 배출총량(TonCo2eq)'] >= 39577.83371812657) & (df['온실가스 배출총량(TonCo2eq)'] < 129813.46140846153),
    (df['온실가스 배출총량(TonCo2eq)'] >= 129813.46140846153) & (df['온실가스 배출총량(TonCo2eq)'] < 301928.6705833333),
    df['온실가스 배출총량(TonCo2eq)'] >= 301928.6705833333
]

# 각 조건에 해당하는 값
values2 = [5, 1, 2, 3, 4]

# 조건에 따라 값을 할당
df['온실가스 배출총량 등급'] = np.select(conditions2, values2)
```
