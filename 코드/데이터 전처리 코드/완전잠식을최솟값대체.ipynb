```python
import pandas as pd
```

```python
df = pd.read_excel('/content/유보율 결측치 채운 것.xlsx')
df.head()
```

```python
# '완전잠식' 값을 가진 컬럼 찾기
columns_with_complete_impairment = [col for col in df.columns if df[col].eq('완전잠식').any()]

# '완전잠식' 값을 제외한 최솟값 구하기
min_values = {}
for col in columns_with_complete_impairment:
    # '완전잠식' 제외한 값들만 필터링
    filtered_values = df[col][df[col] != '완전잠식']
    # 숫자로 변환하여 최솟값 구하기
    numeric_values = pd.to_numeric(filtered_values, errors='coerce')
    if not numeric_values.dropna().empty:
        min_values[col] = numeric_values.min()

# '완전잠식' 값을 최솟값으로 대체하기
for col, min_val in min_values.items():
    df[col] = df[col].replace('완전잠식', min_val)

# 결과 출력
print(df)
```

```python
# '완전잠식' 값이 있는지 확인하기
has_complete_impairment = df.isin(['완전잠식']).any().any()

if has_complete_impairment:
    print("'완전잠식' 값이 데이터프레임에 존재합니다.")
else:
    print("'완전잠식' 값이 데이터프레임에 존재하지 않습니다.")
```

```python
df.to_excel('/content/완전잠식제거.xlsx')
```

```python
# '배당수익률'과 'DPS(원)' 컬럼의 결측치를 0으로 채우기
df['배당수익률'] = df['배당수익률'].fillna(0)
df['DPS(원)'] = df['DPS(원)'].fillna(0)
```
