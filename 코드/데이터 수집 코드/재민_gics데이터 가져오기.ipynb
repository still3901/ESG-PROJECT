```python
import yfinance as yf

# 종목 코드 입력 (예: 'AAPL' for Apple Inc.)
ticker_symbol = "005930.KS"

# yfinance를 통해 해당 종목의 정보를 가져옴
ticker = yf.Ticker(ticker_symbol)

# ticker.info 딕셔너리에서 'sector'와 'industry' 정보를 추출
gics_info = {
    'Sector': ticker.info.get('sector', 'Not Available'),
    'Industry': ticker.info.get('industry', 'Not Available')
}

# GICS 정보 출력
print(f"GICS Sector: {gics_info['Sector']}")
print(f"GICS Industry: {gics_info['Industry']}")
```
